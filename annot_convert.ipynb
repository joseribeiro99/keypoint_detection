```python
import json
import os

def create_annotation_dict(image_annotation, keypoints_annotation, category):
    # Create the shapes list
    shapes = []
    keypoints = keypoints_annotation['keypoints']
    for i in range(0, len(keypoints), 3):
        shapes.append({
            "label": category['name'],
            "points": [[keypoints[i], keypoints[i+1]]],
            "group_id": None,
            "description": "",
            "shape_type": "point",
            "flags": {},
            "mask": None
        })

    # Create the annotation dictionary
    annotation_dict = {
        "version": "5.4.1",
        "flags": {},
        "shapes": shapes,
        "imageHeight": image_annotation['height'],
        "imageWidth": image_annotation['width']
    }

    return annotation_dict

# Create the 'processed_dataset' directory if it doesn't exist
os.makedirs('processed_dataset', exist_ok=True)

# Loop over the annotations
for image_annotation in annotations['images']:
    for keypoints_annotation in annotations['annotations']:
        if image_annotation['id'] == keypoints_annotation['image_id']:
            for category in annotations['categories']:
                if category['id'] == keypoints_annotation['category_id']:
                    # Create the annotation dictionary
                    annotation_dict = create_annotation_dict(image_annotation, keypoints_annotation, category)

                    # Write the annotation dictionary to a json file in the 'processed_dataset' directory
                    with open(os.path.join('processed_dataset', f"{image_annotation['file_name']}_{category['name']}.json"), 'w') as f:
                        json.dump(annotation_dict, f, indent=4)
```
